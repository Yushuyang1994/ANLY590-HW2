Q1

```python
def Relu(x):
    return np.where(x>0, x, 0)

def sigmoid(y):
    return 1/(1+np.exp(-y))

def ff__nn__Relu(x, w, v, b, c):
    z = x.dot(w) + b
    a = Relu(z)
    y_hat = a.dot(v) + c
    return sigmoid(y_hat)
```

```python
W = np.array([[1, -1, 0], [0, 0, .5]])
b = np.array([0, 0, 1])
V = np.transpose(np.array([1, 0, -1]))
c = np.array([1])
X = np.array([[1, -1], [0, -1], [0, 1]])
y_hat = ff__nn__Relu(X, W, V, b, c)
print(y_hat.shape)
print(y_hat)
```

```python
def f(x, y):
    return 1 - x**3 + 100 * (y**2 - x)**2
```

```python
x = np.linspace(-100, 100, 500)
y = np.linspace(-100, 100, 500)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
plt.contour(X, Y, Z, colors='black')
```

```python
def grad_f(x, y):
    dLdx = -3 * x**2 - 200 * (y**2 - x)
    dLdy = 400 * (y**2 - x) * y
    return np.array([dLdx, dLdy])

def grad_descent(x, y, iters, lr):
    values = [x, y]
    trajectory = [values]
    losses = [f(x, y)]
    
    for i in range(iters):
        grad = grad_f(values[0], values[1])
        values = values - lr * grad
        loss = f(values[0], values[1])
        print(values, loss)
        trajectory.append(values)
        losses.append(loss)
    
    return np.array(trajectory), losses
    
```

```python
values, losses = grad_descent(6182, -100, 1000, 0.0000000001)
```

```python
plt.figure(figsize=(8, 5))

plt.subplot(1, 2, 1)
plt.plot(values[:, 0], values[:, 1], '.--')
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(1, 2, 2)
plt.plot(losses)
```
